## Data Cleaning

```python
data.isnull().sum()
```

```python
data[data.isnull().any(axis=1)]
```

```python
data.loc[:122,"Region"]=0
data.loc[122:,"Region"]=1
```

```python
data[['Region']]=data[['Region']].astype('int')
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data[data.isnull().any(axis=1)]
```

```python
data=data.dropna().reset_index(drop=True)
```

```python
data.isnull().sum()
```

```python
data.iloc[[122]]
```

```python
data = data.drop(122).reset_index(drop=True)
```

```python
data.iloc[[122]]
```

```python
data.columns
```

```python
# removing spaces in column names
data.columns=data.columns.str.strip()
```

```python
data.columns
```

```python
# Change Data type
data[['day','month','year','Temperature','RH','Ws']] = data[['day','month','year','Temperature','RH','Ws']].astype('int')
```

```python
data.info()
```

```python
objects = [features for features in data.columns if data[features].dtype=='O']
objects
```

```python
for i in objects:
    if i!= "Classes":
        data[i]=data[i].astype('float')
```

```python
data['Classes']=np.where(data['Classes'].str.contains('not fire'),'not fire','fire')
```

```python
plt.subplots(figsize=(10,6))
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes',data=data)
plt.ylabel('Number of Fire/Non-Fires', weight='bold')
plt.xlabel('Months',weight='bold')
plt.title('Monthly Fire Analysis ')
```

```python
data_region = data.loc[data['Region']==0]
plt.subplots(figsize=(10,6))
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes',data=data_region)
plt.ylabel('Number of Fire/Non-Fires', weight='bold')
plt.xlabel('Months',weight='bold')
plt.title('Fire Analysis of Bejaia Regions')
```
